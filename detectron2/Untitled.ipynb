## Instance Segmentation ##
"It is the task of identifying object outlines at the pixel level."
- __It is identified as one of the hardest possible vision tasks__
A brief overview: 
- Classification: Identifying what's the main context of the image.
- Semantic segmentation: Clearly specifying what we are identifying in the image and focussing on it.
- Object detection: Identifying the number of locations this context is present in this image.
- Instance segmentation: Highlighting the context and pixels that belong to it.

## Mask R-CNN ##

It stands for regional convolutional neural network. 
- Two stage framework:
    + the first stage scans the image and generates proposals (areas likely to contain an object)
    + second stage classifies the proposals and generates boxes and masks.
    
This was introduced in 2018, as an extension to the predecessor, Faster-R-CNN.
- Faster-R-CNN is a popular framework for object detection.
- __Mask-R-CNN__ extends it with instance segmentation, and other things.

### Training Dataset
1. Initial approach should always be searching for public datasets that contain the objects we need. 
2. Based off this article, we can have a smaller dataset if we rely on two main points:
    - Transfer learning: instead of training a model from scratch, we start with a weights file thats been trained on the COCO dataset. This dataset may not have the object we want, but its early learned a lot of the features common in natural images, which really helps.
    - To begin with, we are not setting a high accuracy, so this is a good start!
